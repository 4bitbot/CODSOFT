```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1 = pd.read_csv("/content/Churn_Modelling.csv" )
```

```python
dataset1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
dataset1['Geography'] = label_encoder.fit_transform(dataset1['Geography'])
dataset1['Gender'] = label_encoder.fit_transform(dataset1['Gender'])

```

```python
dataset1.head()
```

```python
dataset1.drop('Surname', axis=1, inplace=True)
```

```python
dataset1.head()
```

```python
correlation_matrix = dataset1.corr()
```

```python
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
dataset1.drop('RowNumber', axis=1, inplace=True)
dataset1.drop('CustomerId', axis=1, inplace=True)


```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the Min-Max scaler
scaler = MinMaxScaler()

# Fit the scaler on the "balance" column (or any other columns you want to scale)
scaler.fit(dataset1[['Balance']])

# Scale the "balance" column
dataset1['Balance'] = scaler.transform(dataset1[['Balance']])

# Reverse the scaling to get the original values
dataset1['Balance'] = scaler.inverse_transform(dataset1[['Balance']])
```
